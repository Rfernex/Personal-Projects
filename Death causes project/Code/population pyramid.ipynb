```python
import pandas as pd # calls the chart
deathcause = pd.read_csv("./cause_of_deaths.csv")
display(deathcause)
```

```python
endcolumn = len(deathcause) #add total death column
deathcause["total deaths"]=deathcause.iloc[:,3:endcolumn].sum(axis=1)
display(deathcause)
```

```python
deathcause.drop(["Country/Territory","Code"], inplace=True, axis=1) #get rid of geographical information
display(deathcause)
```

```python
deathcausebyyear = deathcause.groupby("Year").sum() # groups data by year in rows 
display(deathcausebyyear)
```

```python
deathcausebyyear.reset_index(inplace=True)
display(deathcausebyyear)
```

```python
infectious_diseases_deaths = ["Malaria", "HIV/AIDS", "Tuberculosis", "Lower Respiratory Infections", "Diarrheal Diseases", "Acute Hepatitis"] 
noninfectious_diseases_deaths = ["Alzheimer's Disease and Other Dementias","Parkinson's Disease", "Cardiovascular Diseases", "Neoplasms", "Diabetes Mellitus", "Chronic Kidney Disease", "Chronic Respiratory Diseases", "Cirrhosis and Other Chronic Liver Diseases", "Digestive Diseases"]
idd = deathcausebyyear.loc[:,infectious_diseases_deaths]
nidd = deathcausebyyear.loc[:,noninfectious_diseases_deaths]
#define new subcategories 
```

```python
display(deathcausebyyear)
```

```python
deathcausebyyear["total deaths due to infectious diseases"]=idd.sum(axis=1)
deathcausebyyear["total deaths due to non-infectious diseases"]=nidd.sum(axis=1)
display(deathcausebyyear)
#create columns with total deaths for those categories
```

```python
deathcausebyyear["% deaths due to infectious diseases"]=deathcausebyyear["total deaths due to infectious diseases"]/deathcausebyyear["total deaths"]
deathcausebyyear["% deaths due to non-infectious diseases"]=deathcausebyyear["total deaths due to non-infectious diseases"]/deathcausebyyear["total deaths"]
```

```python
display(deathcausebyyear)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#define x and y limits
y = range(0, len(deathcausebyyear))
x_infectious_death_cause = deathcausebyyear['% deaths due to infectious diseases']
x_noninfectious_death_cause = deathcausebyyear['% deaths due to non-infectious diseases']

#define plot parameters
fig, axes = plt.subplots(ncols=2, sharey=True, figsize=(30, 25))

#specify background color and plot title
fig.patch.set_facecolor('xkcd:light grey')
plt.figtext(.5,.9,"leading death causes pyramid ", fontsize=16, ha='center')
    
#define male and female bars
axes[0].barh(y, x_infectious_death_cause, align='center', color='royalblue')
axes[0].set(title='deaths due to infectious diseases, in % of total deaths')
axes[1].barh(y, x_noninfectious_death_cause, align='center', color='lightpink')
axes[1].set(title='deaths due to non infectious diseases, in % of total deaths')

#adjust grid parameters and specify labels for y-axis
axes[1].grid()
axes[0].set(yticks=y, yticklabels=deathcausebyyear['Year'])
axes[0].invert_xaxis()
axes[0].grid()

#adjust font size
plt.rcParams.update({'font.size': 15})

#display plot
plt.show()


```
